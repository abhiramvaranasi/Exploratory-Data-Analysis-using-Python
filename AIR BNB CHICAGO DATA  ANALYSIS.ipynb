```python
# In case of Asymmetry

# Skewness - Positive skewness indicates peak of curve on left side

# skewness = 0 indicated normal distribution

# Skewness - Negative skewness indicated peak of curve on right side



# Kurtosis - >3 indiacted tall and normal peak

# Kurtosis - =3 indicates Normal Distribution

# Kurtosis - <3 indicatess Wide and Broad peak
```

```python
#  Both nummeric and non Numeric

#  -- Group by Function is used

#  --Leftside of groupby must br numerical

#  --Rightside of grouoby within brackets Non numeric

#  --Statistical Function like sum,mean,median, sd,etc.. must be specidied
```

```python
#  Total number of reviews for top 20 neighbourhoods

chicago.number_of_reviews.groupby(
chicago.neighbourhood).sum().sort_values(
ascending=False).nlargest(10)
```

```python
ax=chicago.number_of_reviews.groupby(
chicago.neighbourhood).sum().sort_values(
ascending=False).nlargest(10).plot(kind='bar',color='red')
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Average price of Top 20 Neighbourhoods

chicago.price.groupby(
chicago.neighbourhood).mean().sort_values(
ascending=False).nlargest(20)
```

```python
plt.figure(figsize=(15,7))
ax=chicago.price.groupby(
chicago.neighbourhood).mean().sort_values(
ascending=False).nlargest(20).plot(kind='bar')
for i in ax.containers:
    ax.bar_label(i)
```

```python
plt.figure(figsize=(15,7))
ax=chicago.price.groupby(
chicago.neighbourhood).mean().sort_values(
ascending=False).nsmallest(20).plot(kind='bar',color='brown')
for i in ax.containers:
    ax.bar_label(i)
```
